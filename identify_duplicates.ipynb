```python
from shapely.geometry import Polygon
import pandas as pd
```

```python
df = pd.read_csv("CAR19_PA_nocanceled.csv")
```

```python
df
```

```python
set(df['year'])
```

```python
landholder_dict = {}
for _, row in df.iterrows():
    landholder_dict[row['cod_imovel']] = row['Landholder']
```

```python
def similar_names(cod_1, cod_2):
    name_1 = landholder_dict[cod_1]
    name_2 = landholder_dict[cod_2]
    if name_1 == name_2:
        return True
    if type(name_1) == float or type(name_2) == float:
        return False
    name_1_set = set((name_1.split())[1:])
    name_2_set = set((name_2.split())[1:])
    if name_1_set.intersection(name_2_set) != set({}):
        return True
    return False
```

```python
coord_dict = {}

for _, row in df.iterrows():
    if row['cod_imovel'] not in coord_dict.keys():
        coord_dict[row['cod_imovel']] = []
    coord_dict[row['cod_imovel']].append((row['X'], row['Y']))

print('Number of properties: ', len(coord_dict.keys()))
```

```python
polygon_dict = {key: Polygon(coord_dict[key]) for key in coord_dict.keys()}
```

```python
df_overlaps = pd.read_csv('CAR21_PA_complete_overlap_table_nobuff.csv', index_col=0)
```

```python
overlapping_dict = {}
for index, row in df_overlaps.iterrows():
    overlapping_dict[index] = [elem for elem in row if elem != 0 and elem != '0']
```

```python
AREA_SIMILARITY = 0.90
OVERLAP_SIMILARITY = 0.70
```

```python
sus_dups = set({})
used = set({})
count = 0
for ID in overlapping_dict.keys():
    if ID in polygon_dict.keys():
        overlapIDs = overlapping_dict[ID]
        pol = polygon_dict[ID]
        for overlapID in overlapIDs:
            if overlapID in polygon_dict.keys():
                o_pol = polygon_dict[overlapID]
                if (abs(pol.area - o_pol.area) < (1-AREA_SIMILARITY)*pol.area and 
                        abs(pol.intersection(o_pol).area - pol.area) < (1-OVERLAP_SIMILARITY)*pol.area and
                            similar_names(ID, overlapID)):
                    if ID not in used:
                        sus_dups.add(ID)
                    used.add(overlapID)
        count += 1
        if count % 20000 == 0:
            print(count)
```

```python
print('Number of suspected duplicates: ', len(sus_dups))
```

```python
list(sus_dups)[0]
```

```python
dup_dict = {}
for key in polygon_dict.keys():
    if key in sus_dups:
        dup_dict[key] = [1]
    else:
        dup_dict[key] = [0]
```

```python
df_out = pd.DataFrame.from_dict(dup_dict, orient='index', columns=['is_duplicate'])
```

```python
compression_opts = dict(method='zip', archive_name='CAR19_PA_dups_pass1.csv')
df_out.to_csv('CAR19_PA_dups_pass1.zip', index=True, compression=compression_opts)
```
